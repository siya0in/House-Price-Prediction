```python
X = data.drop(columns=['price', 'price_per_sqft'])
y = data[['price', 'price_per_sqft']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_selector
from sklearn.pipeline import make_pipeline
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

Applying Linear Regression 

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

y = y.drop(columns=['price_per_sqft'])
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

# Define your column transformer
column_trans = make_column_transformer((OneHotEncoder(), ['beds']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# X being feature matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression()
lr.fit(X_scaled, y)
```

```python
pipe = make_pipeline(column_trans,scaler, lr)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

# Remove the step related to 'beds' from your column transformer
column_trans = make_column_transformer((OneHotEncoder(), ['bathroom']), remainder='passthrough')

# Update your pipeline to exclude the step related to 'beds'
pipe = make_pipeline(column_trans, StandardScaler(), LinearRegression())

# Fit the pipeline
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred_lr)
print("R^2 Score:", r2)
```

Using Lasso 

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

Using Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test,y_pred_lr))
print("Lasso: ", r2_score(y_test,y_pred_lasso))
print("Ridge: ", r2_score(y_test,y_pred_ridge))
```
